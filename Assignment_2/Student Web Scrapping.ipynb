```python
quotations = []

for pages in range(1,5):
    url = f"http://quotes.toscrape.com/page/{pages}/"
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response,'html.parser')  
    div = soup.find('div')
    items = div.find_all('div', class_ = 'quote')
    
    for item in items:
        name = item.find('small', class_ = 'author').text
        quote = item.find('span', class_ = 'text').text
        tags = [tag.text for tag in item.find_all('a', class_ = 'tag')]
        quotations.append([name,quote,tags])
```

```python
df = pd.DataFrame(quotations,columns = ['Name','Quote','Tags'])
```

```python
df.head()
```

```python
df['Quote'] = df['Quote'].str.strip('“”')
```
